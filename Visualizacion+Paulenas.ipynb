```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/Ipaulenas/CoderHouse/main/Telco_customer_churn_services.csv"
df1 = pd.read_csv (url)
```

```python
mpl.style.use ('bmh')  #Estilo bmh
```

```python
df1
```

```python
y = df1['Contract'].value_counts()
fig, ax = plt.subplots ()
ax.bar (y.index , y )
ax.set_title ("Cantidad de clientes por los tipos de contratos")
ax.set_xlabel('Tipos de contrantos')
ax.set_ylabel('Cantidad de clientes')
```

```python
y = df1.groupby(['Contract'])['Total Revenue'].sum()
fig, ax = plt.subplots ()
ax.plot (y.index , y )
#ax.set_ylim (0 , 10000)
ax.set_title ("Ganancias por los tipos de contratos")
ax.set_xlabel('Tipos de contrantos')
ax.set_ylabel('Ganancias por cada tipo de contrato')
```

```python
servicio_telefonico = df1['Phone Service'].value_counts()
multiple_linea = df1['Multiple Lines'].value_counts()
servicio_internet = df1['Internet Service'].value_counts()
servicio_tv = df1['Streaming TV'].value_counts()

fig, ax = plt.subplots ()
ax.bar (servicio_telefonico.index , servicio_telefonico , label ='Servicio telefonico',  alpha = 0.5 )
ax.bar (multiple_linea.index , multiple_linea , label ='Multiple linea' , color = 'r',  alpha = 0.5)
ax.bar (servicio_internet.index , servicio_internet , label ='Servicio internet' , color = 'g' , alpha = 0.5)
ax.bar (servicio_tv.index , servicio_tv , label ='Servicio TV' , color = 'b', alpha = 0.5)

ax.set_title ("Tipos de servicios contratados")
ax.set_xlabel('Tiene contratado el servicio')
ax.set_ylabel('Cantidad de clientes que contrataron los servicios')
ax.legend ()
```

```python
import pandas as pd
import seaborn as sns
tipo_contrtato = "https://raw.githubusercontent.com/Ipaulenas/CoderHouse/main/Telco_customer_churn_services.csv"
df1 = pd.read_csv (tipo_contrtato)

estado_del_servicio = "https://raw.githubusercontent.com/Ipaulenas/CoderHouse/main/Telco_customer_churn_status.csv"
df2 = pd.read_csv (estado_del_servicio)

datos_clientes = "https://raw.githubusercontent.com/Ipaulenas/CoderHouse/main/Telco_customer_churn_demographics.csv"
df3 = pd.read_csv (datos_clientes)
```

```python
import pandas as pd
# Unimos los dos data sets en función de la columna “Customer ID”
df_final = pd.merge(df3, df2, on='Customer ID')
# Imprimimos el data set final
df_final

```

```python
df_final = df_final.replace ({"F": "Female" , "M": 'Male'})
```

```python
cant = df_final['Gender'].value_counts()
print (cant)
```

```python
#Axes.level
#df.loc[condicion.columnas]
#df.loc[penguins.species == 'Adeline',:]
#sns.histplot(datafram, x = un arreglo o pandas series, hue = division por categoria, multiple = tipo de grafico historgrafico)
#sns.histplot(data= penguins.loc [], x = "flipper_le", multiple = "stack")

cant = df_final['Customer Status'].value_counts()
print (cant)
g = sns.barplot(data = df_final, x = cant.index, y =  cant  ) 
g.set_title("Cantidad de clientes por estado de los mismos")
g.set_xlabel('Estado de los clientes')
g.set_ylabel('Cantidad de clientes')
```

```python
cant = df_final['Customer Status']
cant = cant.loc[df_final['Customer Status'] == "Churned"]
cant2 = df_final['Age'].value_counts ()
f, g = plt.subplots(figsize=(6, 6))

g = sns.lineplot(data = df_final, x = df_final['Age'] ,  y = cant2 , hue = cant ) 
g.set_title("Clientes que descontraron el servicio por edad")
g.set_xlabel('Edad de los clientes')
g.set_ylabel('Cantidad de clientes')
```
